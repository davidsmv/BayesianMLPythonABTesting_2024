```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
mu = 170
sd = 7

x = norm.rvs(loc=mu, scale=sd, size=100)
```

```python
# Creating the histogram
plt.hist(x, bins=15, density=True, alpha=0.6, color='g')

# Plotting the PDF of the normal distribution
xmin, xmax = plt.xlim()
x_lin = np.linspace(xmin, xmax, 100)
p = norm.pdf(x_lin, mu, sd)
plt.plot(x_lin, p, 'k', linewidth=2)
title = "Fit results: mu = %.2f,  std = %.2f" % (mu, sd)
plt.title(title)
plt.show()
```

```python
# maximun likelihood mean
x.mean()
```

```python
# maximun likelihood variance
x.var()
```

```python
# maximun likelihood std
x.std()
```

```python
# unbiased variance
x.var(ddof=1)
```

```python
# unbiased std
x.std(ddof=1)
```

```python
# at what height are you in the 95th percentile?
norm.ppf(0.95, loc=mu, scale=sd)
```

```python
# you are 160 cm tall, what percentile are you in?
norm.cdf(160, loc=mu, scale=sd)
```

```python
# you are 180 cm tall, what is the probability that someone is taller than you?
1 - norm.cdf(180, loc=mu, scale=sd)
```
